```python
ratingwithbooks = rating.merge(book,on='ISBN')
```

```python
number_rating = ratingwithbooks.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating.rename(columns = {'Book-Rating':'number of ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratingwithbooks.merge(number_rating,on='Book-Title')
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['User-ID','Book-Title'],inplace=True)
```

```python
final_rating.shape
```

```python
#now we have to creat a pivot table

book_pivot = final_rating.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
bookspr = csr_matrix(book_pivot)
```

```python
type(bookspr)
```

```python
bookspr
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(bookspr)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1))
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
    
```

```python
# what if we only want to give book title and want recommendations
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommendBooks(bookname):
    bookid=np.where(book_pivot.index==bookname)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[bookid,:].values.reshape(1,-1))
    
    for i in range(len(suggestions)):
        print(book_pivot.index[suggestions[i]])
```

```python
recommendBooks('4 Blondes')
```

```python
recommendBooks('Zoya')
```
